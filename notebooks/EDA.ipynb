### Outliers removal
Outliers are nothing but the values that are rare and out of the specific range,(box plot itself decides the range and shows the values/points lying outside)

```python
sns.boxplot(df['trestbps'])
```

```python
df = df.drop(df[(df['trestbps']>170)].index)
sns.boxplot(df['trestbps'])
```

```python
sns.boxplot(df['chol'])
```

```python
df = df.drop(df[(df['chol']>350)].index)
sns.boxplot(df['chol'])
```

```python
sns.boxplot(df['thalach'])
```

```python
df = df.drop(df[(df['thalach']<80)].index)
sns.boxplot(df['thalach'])
```

```python
sns.boxplot(df['oldpeak'])
```

```python
df = df.drop(df[(df['oldpeak']>4)].index)
sns.boxplot(df['oldpeak'])
```

```python
plt.figure(figsize=(10,12))
sns.heatmap(df.corr(), annot=True, robust=True)
```

   ### Observations
   Columns having correlation of more than 0.35 should be considered as a strong relation.
   The columns having more than specified corr are
   * Age and ca
   * Sex and thal
   * cp and exang, cp and condition
   * exang and condition
   * old peak and slope, old peak and condition
   * ca and condition
   * thal and condition

```python
sns.barplot(x=df['ca'], y=df['age'], data=df)
```

We can see that with increasing age, it looks more and more like you have a type 2 and type 3 coronary calcium scan, which has a strong correlation with this disease. The CA is mainly observed in the age group over 50 years.

```python
sns.countplot(df['thal'], hue=df['sex'])
```

* 0: can be seen in both male and female
* 1: can be seen for more male than in female
* 2: can be seen in most of the male and very less in female
